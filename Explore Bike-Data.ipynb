```python
import time
import pandas as pd
import numpy as np
```

```python

Df_1 = pd.read_csv('chicago.csv').assign(City='chicago').set_index('City')
Df_2 = pd.read_csv('new_york_city.csv').assign(City='newyork').set_index('City')
Df_3 = pd.read_csv('washington.csv').assign(City='washington').set_index('City')
```

```python
df=pd.concat([Df_1, Df_2,Df_3])
```

```python
df
```

```python
df["Start Time"] = pd.to_datetime(df["Start Time"], format="%Y-%m-%d %H:%M")
df["End Time"] = pd.to_datetime(df["Start Time"], format="%Y-%m-%d %H:%M")
```

```python
start_time = time.time()

df['Month'] = df['Start Time'].dt.strftime('%m')
df['Hour'] = df['Start Time'].dt.strftime('%H')
df['DayofWeak']=df['Start Time'].dt.day_name()
df['Day'] = df['Start Time'].dt.day

print("\nThis took %s seconds." % (time.time() - start_time))
```

```python
def get_filters():  
    
    
    Flag= True
    while Flag:   
        City=input('Put your City: ').lower()  
        if City not in ['chicago','newyork','washington','none']:   
            print('city should be (chicago, newyork, washington) or "None" \n')
        else:  
            Flag= False
      
    print('▬'*40)
    
    
    
    Flag= True
    while Flag:   
        Month=input('Put your Month: ').lower()
        if Month == "none":
            break    
        if Month in ['01','02','03','04','05','06']: 
            Flag= False
        else:  
            print('Month should be num between [01:06] like 05 \n')
                    
    print('▬'*40)
    
    Flag= True
    while Flag: 
        
        Day=input('Put your Day: ' ).lower()
        if Day == "none":
            break 
        Day=int(Day)  
        if 1<= Day <=30:   
            Flag= False
        else:  
            print('Day should be num \n')       
                   
        
    print('▬'*40)
    return City,Month,Day



```
